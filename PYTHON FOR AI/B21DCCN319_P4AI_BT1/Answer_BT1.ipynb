```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
```

```python
# Đọc dữ liệu từ file

file_path = 'P4AI_BT1.csv'
data = pd.read_csv(file_path)
```

```python
# Hiển thị toàn bộ dữ liệu
print("Hiển thị toàn bộ dữ liệu:")
display(data)
```

```python
# Điền các giá trị bị thiếu
data['sepal.length'].fillna(data['sepal.length'].mean(), inplace=True)
data['sepal.width'].fillna(data['sepal.width'].mean(), inplace=True)
data['petal.length'].fillna(data['petal.length'].mean(), inplace=True)
data['petal.width'].fillna(data['petal.width'].mean(), inplace=True)
data['variety'].fillna(data['variety'].mode()[0], inplace=True)
```

```python
# Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3
filtered_data = data[(data['sepal.length'] > 5) & (data['sepal.width'] > 3)]
print("Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3:")
display(filtered_data)
```

```python
# Chuẩn hóa dữ liệu theo min max
scaler = MinMaxScaler()
normalized_data = data.copy()
normalized_data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']] = scaler.fit_transform(
    normalized_data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
)
print("Chuẩn hóa dữ liệu theo min max:")
display(normalized_data)
```

```python
# Liên tục hóa thuộc tính variety
label_encoder = LabelEncoder()
continuous_data = data.copy()
continuous_data['variety'] = label_encoder.fit_transform(continuous_data['variety'])
print("Liên tục hóa thuộc tính variety:")
display(continuous_data)
```

```python
# Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety dựa trên các thuộc tính khác
X = continuous_data.drop('variety', axis=1)
y = continuous_data['variety']

model = LinearRegression()
model.fit(X, y)
```

```python
model.predict(X)
```

```python
# Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu
sampled_data = data.sample(frac=0.5, replace=True, random_state=1)
print("Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu:")
display(sampled_data)
```

```python
# Rời rạc hóa thuộc tính sepal.length
data['sepal.length.discrete'] = pd.cut(data['sepal.length'], bins=3, labels=False)
print("Rời rạc hóa thuộc tính sepal.length:")
display(data)
```

```python
# Phân cụm dữ liệu và loại bỏ ngoại lai
# Giả sử k=3 cho phân cụm
kmeans = KMeans(n_clusters=3, random_state=1)
data['cluster'] = kmeans.fit_predict(data.drop('variety', axis=1))
```

```python
# Loại bỏ ngoại lai sử dụng Isolation Forest
iso_forest = IsolationForest(contamination=0.1, random_state=1)
data['outlier'] = iso_forest.fit_predict(data.drop(['variety', 'cluster'], axis=1))
data_without_outliers = data[data['outlier'] == 1]
print("Phân cụm dữ liệu và loại bỏ ngoại lai:")
display(data_without_outliers)
```
