### **BÀI TẬP BUỔI 2: TIỀN XỬ LÝ DỮ LIỆU**



**Câu 1:** Hiển thị toàn bộ dữ liệu

```python
import pandas as pd
data = pd.read_csv('P4AI_BT1.csv')
display(data)
```

**Câu 2.** Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc

```python
df1 = data.copy()
x = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width', 'variety']
for column in x:
    if df1[column].dtype == 'object':  df1[column].fillna(df1[column].mode()[0], inplace=True)
    else: df1[column].fillna(round(df1[column].mean(), 2), inplace=True)
display(df1)
```

**Câu 3.** Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3

```python
df = df1.copy()
df = df[(df['sepal.length'] > 5) & (df['sepal.width'] > 3)]
display(df)
print()
```

**Câu 4.** Chuẩn hóa dữ liệu theo min max

```python
df3 = df1.copy()
x = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
for y in x: df3[y] = (df3[y] - df3[y].min())/(df3[y].max() - df3[y].min())
display(df3)
```

**Câu 5.** Liên tục hoá trường variety

```python
from sklearn.preprocessing import LabelEncoder
df4 = df1.copy()
label_encoder = LabelEncoder()
df4['variety'] = label_encoder.fit_transform(df4['variety'])
display(df4)
```

**Câu 6.** Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính *variety* dựa trên các thuộc tính khác

```python
from sklearn.linear_model import LinearRegression
x = df4.drop('variety', axis=1) #x: đầu vào chứa mọi cột nhưng bỏ variety
y = df4['variety'] #y: đầu ra
model = LinearRegression()
model.fit(x, y)
```

```python
model.predict(x)
```

**Câu 7.**Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu

```python
sampled_data =df1.sample(frac=0.5, replace=True)
display(sampled_data)
```

**Câu 8.** Rời rạc hoá trường sepal.length

```python
df5 = df3.copy()
df5['sepal.length'] = pd.cut(df5['sepal.length'], bins = [0, 0.25, 0.5, 0.75, 1], labels = ['Low', 'Medium', 'High', 'Very High'])
display(df5)
```

**Câu 9.** Phân cụm dữ liệu và loại bỏ ngoại lai

**a.** Phân cụm dữ liệu


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=1)
df1['cluster'] = kmeans.fit_predict(df1.drop('variety', axis=1)) #cluster: số nhóm, cho biết data ấy thuộc nhóm nào
```

**b.** Loại bỏ ngoại lai

```python
from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination=0.1, random_state=1) #Ước tính có 10% dữ liệu ngoại lai
df1['outlier'] = iso_forest.fit_predict(df1.drop(['variety', 'cluster'], axis=1)) #Loại bỏ các cột không cần thiết
data_without_outliers = df1[df1['outlier'] == 1] #1: Không ngoại lai, -1: Ngoại lai
display(data_without_outliers)
```
